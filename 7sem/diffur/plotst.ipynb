```python
import matplotlib.pyplot as plt


with open("data.txt", "r") as f:
    data = [float(line.strip()) for line in f]

even_data = []
odd_data = []
for i in range(0, len(data), 2):
    
    even_data.append(data[i])
    odd_data.append(data[i+1])






plt.plot(even_data, odd_data)
plt.show()
```
