```python
import matplotlib.pyplot as plt
import numpy as np


with open("data.txt", "r") as f:
    data = [list(map(float, line.split())) for line in f]


x, y, t = zip(*data)
# print(np.diff(t))
print(f"average step: {np.mean(np.diff(t))}")
plt.plot(t, x)
plt.show()

plt.plot(t, y)
plt.show()

plt.plot(x, y)
plt.show()

plt.plot(np.diff(t), t)
plt.show()
print(len(x))
```
