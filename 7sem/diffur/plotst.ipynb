```python
import matplotlib.pyplot as plt
import numpy as np


with open("data.txt", "r") as f:
    data = [list(map(float, line.split())) for line in f]
with open("errorLog.txt", "r") as err:
    errorData = [list(map(float, line.split())) for line in err]
with open("errorRegular.txt", "r") as errReg:
    errorRegData = [list(map(float, line.split())) for line in errReg]

x, y, t = zip(*data)
error, logNorm, time = zip(*errorData)
errorReg, normReg, timeReg = zip(*errorRegData)

# print(np.diff(t))
print(f"average step: {np.mean(np.diff(t))}")
plt.plot(t, x)
plt.show()

plt.plot(t, y)
plt.show()

plt.plot(x, y)
plt.show()

plt.plot(time, error)
plt.show()

plt.plot(time, logNorm)
plt.show()

plt.plot(time, errorReg)
plt.show()

plt.plot(time, normReg)
plt.show()

print(len(x))
```
